```python
import numpy as np
import torch
```

```python
filepath = "../pytorch_dlwpt-code-master/data/p1ch4/bike-sharing-dataset/hour-fixed.csv"
bikes_numpy = np.loadtxt(
    filepath,
    dtype=np.float32,
    delimiter=",",
    skiprows=1,
    converters={1: lambda x: float(x[8:10])},
)
bikes_numpy
```

```python
bikes = torch.from_numpy(bikes_numpy)
bikes
```

```python
bikes.shape, bikes.stride()
```

```python
bikes.shape[1]
```

```python
daily_bikes = bikes.view(-1, 24, bikes.shape[1])
daily_bikes
```

The goal of this code is to reshape the bikes tensor into a new 3D tensor where:

The first dimension corresponds to the number of days (calculated automatically).
The second dimension is 24, representing 24 hours per day.
The third dimension is the number of columns/features (e.g., data like the number of bikes sold or temperature).


```python
daily_bikes.shape, daily_bikes.stride()
```

```python
daily_bikes = daily_bikes.transpose(1, 2)
daily_bikes.shape, daily_bikes.stride()
daily_bikes
```

```python
first_day = bikes[:24].long()
whether_onehot = torch.zeros(first_day.shape[0], 4)
first_day[:, 9]
```

```python
first_day = bikes[:24].long()
first_day
```

```python
whether_onehot = torch.zeros(first_day.shape[0], 4)
whether_onehot.shape
```

```python
whether_onehot.scatter_(dim=1, index=first_day[:, 9].unsqueeze(1).long() - 1, value=1.0)
```

```python
daily_weather_onehot = torch.zeros(daily_bikes.shape[0], 4, daily_bikes.shape[2])
daily_weather_onehot.shape
```

```python
daily_weather_onehot.scatter_(1, daily_bikes[:, 9, :].long().unsqueeze(1) - 1, 1.0)
daily_weather_onehot.shape
```

```python
daily_bikes = torch.cat((daily_bikes, daily_weather_onehot), dim=1)
```

```python
daily_bikes[:, 9, :] = (daily_bikes[:, 9, :] - 1.0) / 3.0
```

```python
temp = daily_bikes[:, 10, :]
temp_min = torch.min(temp)
temp_max = torch.max(temp)
daily_bikes[:, 10, :] = (daily_bikes[:, 10, :] - temp_min) / (temp_max - temp_min)
```

```python
temp = daily_bikes[:, 10, :]
daily_bikes[:, 10, :] = (daily_bikes[:, 10, :] - torch.mean(temp)) / torch.std(temp)
```
