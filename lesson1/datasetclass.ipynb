```python
from torch.utils.data import Dataset
import cv2
import numpy as np
import os


class Datasetclass(Dataset):
    def __init__(self, filepath, filedir) -> None:
        self.path = filepath
        self.filedir = filedir
        self.img_dir = os.path.join(self.path, self.filedir)
        self.img_list = os.listdir(self.img_dir)

    def __getitem__(self, index):
        self.img_name = self.img_list[index]
        self.img = os.path.join(self.path, self.filedir, self.img_name)
        self.label = self.img_name

        return self.img, self.label

    def __len__(self):
        return len(self.img_list)


path = "./data/hymenoptera_data/train/"
filedir = "ants"
data = Datasetclass(path, filedir)

print(data.__len__())
print(data.__getitem__(0))
imgpath = data.__getitem__(0)[0]
img = cv2.imread(imgpath)
# Display the image
cv2.imshow("Image", img)
cv2.destroyAllWindows()
# Wait until a key is pressed, then close the window
cv2.waitKey(0)
```
