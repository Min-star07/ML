```python
import torch
import torchvision
from torch.utils.data import DataLoader

dataset = torchvision.datasets.CIFAR10(
    "../lesson3/dataset",
    train=False,
    transform=torchvision.transforms.ToTensor(),
    download=True,
)

dataloader = DataLoader(dataset, batch_size=64)


class NET(torch.nn.Module):
    def __init__(self):
        super(NET, self).__init__()
        self.conv1 = torch.nn.Conv2d(
            in_channels=3, out_channels=6, kernel_size=3, stride=1, padding=0
        )

    def forward(self, x):
        return self.conv1(x)


model = NET()
print(model)

for data in dataloader:
    inputs, targets = data
    output = model(inputs)
    print(output.shape)
```

```python
# maxpool
class NET(torch.nn.Module):
    def __init__(self):
        super(NET, self).__init__()
        self.maxpool = torch.nn.MaxPool2d(kernel_size=3)

    def forward(self, x):
        return self.maxpool(x)


model = NET()
print(model)

input = [3, 4, 6, 5, 7, 2, 4, 6, 8, 2, 1, 6, 7, 8, 4, 9, 7, 4, 6, 2, 3, 7, 5, 4, 1]
input = torch.Tensor(input).view(-1, 5)
input = torch.reshape(input, (1, 1, 5, 5))

print(input)
output = model(input)
print(output)
```

```python
import torch
import torchvision
from torch.utils.data import DataLoader

dataset = torchvision.datasets.CIFAR10(
    "../lesson3/dataset",
    train=False,
    transform=torchvision.transforms.ToTensor(),
    download=True,
)

dataloader = DataLoader(dataset, batch_size=64)


class NET(torch.nn.Module):
    def __init__(self):
        super(NET, self).__init__()
        self.conv1 = torch.nn.Conv2d(3, 32, 5, stride=1, padding=2)
        self.maxpool1 = torch.nn.MaxPool2d(2)
        self.conv2 = torch.nn.Conv2d(32, 32, 5, stride=1, padding=2)
        self.maxpool2 = torch.nn.MaxPool2d(2)
        self.conv3 = torch.nn.Conv2d(32, 64, 5, stride=1, padding=2)
        self.maxpool3 = torch.nn.MaxPool2d(2)
        self.flatten = torch.nn.Flatten()
        self.l1 = torch.nn.Linear(1024, 64)
        self.l2 = torch.nn.Linear(64, 10)

    def forward(self, x):
        x = self.conv1(x)
        x = self.maxpool1(x)
        x = self.conv2(x)
        x = self.maxpool2(x)
        x = self.conv3(x)
        x = self.maxpool3(x)
        x = self.flatten(x)
        x = self.l1(x)
        x = self.l2(x)

        return x


model = NET()
print(model)

input = torch.ones((64, 3, 32, 32))
output = model(input)
print(output.shape)
# for data in dataloader:
#     inputs, targets = data
#     output = model(inputs)
#     print(output.shape)
```

```python
import torch
import torchvision
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter

dataset = torchvision.datasets.CIFAR10(
    "../lesson3/dataset",
    train=False,
    transform=torchvision.transforms.ToTensor(),
    download=True,
)

dataloader = DataLoader(dataset, batch_size=64)


class NET(torch.nn.Module):
    def __init__(self):
        super(NET, self).__init__()
        self.model1 = torch.nn.Sequential(
            torch.nn.Conv2d(3, 32, 5, stride=1, padding=2),
            torch.nn.MaxPool2d(2),
            torch.nn.Conv2d(32, 32, 5, stride=1, padding=2),
            torch.nn.MaxPool2d(2),
            torch.nn.Conv2d(32, 64, 5, stride=1, padding=2),
            torch.nn.MaxPool2d(2),
            torch.nn.Flatten(),
            torch.nn.Linear(1024, 64),
            torch.nn.Linear(64, 10),
        )

    def forward(self, x):
        x = self.model1(x)
        return x


model = NET()
print(model)

input = torch.ones((64, 3, 32, 32))
output = model(input)
print(output.shape)
writer = SummaryWriter("./log_seq")
writer.add_graph(model, input)
writer.close()
```
